```python
import pandas as pd
import matplotlib.pyplot as plt
from brokenaxes import brokenaxes
import numpy as np

tlb_latency = 80 #cycles
```

# Figure 2

Assume dTLB-load-misses has the same level latency of LLC-misses: ~80cycles.

```python
df_10m_dense = pd.read_csv('art-10m-sortedkey-dense.csv')
dense_ratio = (tlb_latency * df_10m_dense['dTLB-load-misses']) / df_10m_dense['cycles']

df_10m_sparse = pd.read_csv('art-10m-sparse.csv')
sparse_ratio = (tlb_latency * df_10m_sparse['dTLB-load-misses']) / df_10m_sparse['cycles']

plt.plot(df_10m_dense['alpha'], dense_ratio, label='Dense')
plt.plot(df_10m_dense['alpha'], sparse_ratio, label='Sparse')
plt.legend()
plt.xlabel("zipf factor")
plt.ylabel("tlb stall time / lookup latency")
plt.title("tlb stall time / lookup latency in % -- ART 10M keys")
plt.show()
```
