```python
import pandas as pd
import matplotlib.pyplot as plt
from brokenaxes import brokenaxes
import numpy as np
```

# Pure-ART Dense

```python
df = pd.read_csv('art-1m-sortedkey-dense.csv')
# plt.plot(df['alpha'], df['throughput'])
# plt.plot(df['alpha'], df['cycles'])
plt.plot(df['alpha'], df['L1-misses'], label='L1-misses')
plt.plot(df['alpha'], df['LLC-misses'], label='LLC-misses')
plt.plot(df['alpha'], df['dTLB-load-misses'], label='dTLB-load-misses')
plt.legend()
plt.xlabel("zipf factor")
plt.ylabel("counter: #")
plt.title("ART 1M Key Dense Sorted")
plt.show()
```
