```python
import pandas as pd
import matplotlib.pyplot as plt
from brokenaxes import brokenaxes
import numpy as np

tlb_latency = 80 #cycles
```

# Figure 2

Assume dTLB-load-misses has the same level latency of LLC-misses: ~80cycles.

```python
df_10m_dense = pd.read_csv('art-10m-sortedkey-dense.csv')
dense_ratio = 100 * (tlb_latency * df_10m_dense['dTLB-load-misses']) / df_10m_dense['cycles']

df_10m_sparse = pd.read_csv('art-10m-sparse.csv')
sparse_ratio = 100 * (tlb_latency * df_10m_sparse['dTLB-load-misses']) / df_10m_sparse['cycles']

plt.plot(df_10m_dense['alpha'], dense_ratio, label='Dense', color='r')
plt.plot(df_10m_dense['alpha'], sparse_ratio, label='Sparse', color='g')
plt.legend()
plt.xlabel("zipf factor")
plt.ylabel("tlb stall time / lookup latency")
plt.title("tlb stall time / lookup latency in % -- ART 10M keys")
plt.show()
```

# Figure 3


```python
df_10m_dense = pd.read_csv('art-10m-sortedkey-dense.csv')
df_10m_sparse = pd.read_csv('art-10m-sparse.csv')

df_10m_dense_huge = pd.read_csv('huge-art-10m-sortedkey-dense.csv')
dense_ratio = 100 * (df_10m_dense_huge['throughput'] / df_10m_dense['throughput'] - 1.0)

df_10m_sparse_huge = pd.read_csv('huge-art-10m-sparse.csv')
sparse_ratio = 100 * (df_10m_sparse_huge['throughput'] / df_10m_sparse['throughput'] - 1.0)

plt.plot(df_10m_dense['alpha'], dense_ratio, label='Dense', color='r')
plt.plot(df_10m_dense['alpha'], sparse_ratio, label='Sparse', color='g')
plt.legend()
plt.xlabel("zipf factor")
plt.ylabel("throughput improvement (%)")
plt.title("Throughput Improvement by using Huge Page over Regular Page")
plt.show()
```
